# <center> EDA </center>

```python
import numpy as np
import pandas as pd
from ast import literal_eval
```

# Movies

```python
movies = pd.read_csv(r"data/top_movies_clean.csv",converters={"genres": literal_eval,"top_cast": literal_eval,"writers": literal_eval,"plot_tags": literal_eval}) 
display(movies.shape)
movies.sort_values('total_ratings',ascending=False).head(3)
```

# <center> -------------------------------------- </center>

# 1) Ratings:
- Negative skew
- Most movies between 5.6 - 7.6, mean 6.6

```python
display(movies['rating'].describe())
movies['rating'].hist();
```

# 2) Dates:
- Negative Skew towards more recent dates

```python
display(movies['date'].describe())
movies['date'].hist();
```

# 3) Genres:

### Counts

```python
movies.explode('genres')['genres'].value_counts().plot(kind='bar', figsize=(12,5));
```

```python
movies[['genres','rating']].explode('genres').boxplot(column='rating', by='genres', figsize=(20,6));
```

# 4) Summaries:

```python
movies['summaries'].apply(lambda x: len(x.split())).hist();
```
